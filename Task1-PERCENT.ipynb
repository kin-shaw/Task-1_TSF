## Step-2 : Exploring the data
### Summarizing Data

```r
str(percent)
summary(percent)
```

### Plotting Percentage v/s Hours

```r
plot(percent$Hours,percent$Percent,main="Percentage v/s Hours",xlab="Hours",ylab="Percentage")
```

### Frequency Chart for No. of Hours and Percentage

```r
hist(percent$Hours)
hist(percent$Percent)
```

### Correlation Matrix : Gives how strongly study hours determine percentage on scale of 0 to 1

```r
cor(percent[c("Hours", "Percent")])
cor(percent$Hours,percent$Percent)
```

```r
pairs(percent[c("Hours", "Percent")])
```

### Summary of Hours and Percentage

```r
pairs.panels(percent[c("Hours", "Percent")])
```

## Step -3 : Preparing train and test data

```r
percent_train=percent[1:22,]
percent_test=percent[23:25,]
```

## Step-4 : Applying Linear Regression

```r
p_model= lm(Percent ~ Hours, data = percent_train)
plot(p_model)
```

### Predicted v/s Actual values for Test data

```r
table(predict(p_model,newdata=percent_test),percent[23:25,2:2])
```

## Step-5 : Evaluating model performance

```r
summary(p_model)
```

## Predicting Percentage of student studying for 9.25 hours

```r
p=as.data.frame(9.25)
colnames(p)="Hours"
```

```r
predict(p_model,newdata=p)
```
